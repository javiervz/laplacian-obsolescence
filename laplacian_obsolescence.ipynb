## read data from glottolog!

```python
import pandas as pd
```

```python
# https://glottolog.org/meta/downloads

df = pd.read_csv('languoid.csv',sep=',')
```

```python
df
```

```python
# languages with iso code

df = df[pd.notnull(df['iso639P3code'])]
```

```python
# languages with level "language" or "dialect"

df = df[(df['level'] == 'language')|(df['level'] == 'dialect')]
```

```python
df
```

```python
df_family = df.groupby('family_id')['iso639P3code'].apply(list).reset_index(name='languages')
```

```python
# dictionary number:family
families = df_family.to_dict()['family_id']

#dictionary family_name:list of iso codes
language_families = {families[key]:df_family.to_dict()['languages'][key] for key in df_family.to_dict()['languages'].keys()}
```
